```python
df = pd.DataFrame(source_document)
df.columns = ["Title", "Content"]
df
```

```python
def embed_fn(title, text) :
  return genai.embed_content(
    model=model,
    content=text,
    title=title,
    task_type="RETRIEVAL_DOCUMENT")["embedding"]
```

```python
df["Embeddings"] = df.apply(lambda row: embed_fn(row["Title"], row["Content"]), axis=1)
df
```

```python
model = "models/embedding-001"
```

```python
def gen_and_fetch(query, base, model) :
  gen_and_fetch_embeding = genai.embed_content(
    model=model,
    content=query,
    task_type="RETRIEVAL_QUERY")["embedding"]

  scalar_product = np.dot(np.stack(df["Embeddings"]), gen_and_fetch_embeding)

  indx = np.argmax(scalar_product)
  return df.iloc[indx]["Content"]
```

```python
def break_lines(text, width=150):
    return '\n'.join(text[i:i+width] for i in range(0, len(text), width))
```

```python
prompt = input("Descreva a sua desconfiança de golpe: ")

while prompt != "end":
  response = gen_and_fetch(prompt, df, model)
  formatted_res = break_lines(response)
  print(formatted_res + "\n")
  prompt = input("Descreva a sua desconfiança de golpe: ")
```
